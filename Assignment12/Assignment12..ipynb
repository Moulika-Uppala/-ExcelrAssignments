```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Count missing values
missing_count = df.isnull().sum()
print("\nMissing values per column:\n", missing_count)

# Percentage missing
missing_percent = (df.isnull().mean() * 100).round(2)
print("\nMissing values percentage:\n", missing_percent)

# Visualize missing values
df.isnull().sum().plot(kind='bar', figsize=(10,5), title="Missing Values per Column")
plt.show()

# OUTLIER DETECTION (IQR METHOD)

print("\nOUTLIERS (IQR) : ")

def detect_outliers_iqr(data):
    outlier_indices = []
    numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
    
    for col in numeric_cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = data[(data[col] < lower) | (data[col] > upper)].index
        print(f"{col}: {len(outliers)} outliers")

        outlier_indices.extend(outliers)

    return list(set(outlier_indices))

iqr_outliers = detect_outliers_iqr(df)
print("Total rows with outliers:", len(iqr_outliers))

# 4. OUTLIER DETECTION (Z-SCORE METHOD)

print(" OUTLIERS (Z-SCORE) : ")

def detect_outliers_zscore(data, threshold=3):
    numeric_cols = data.select_dtypes(include=['float64', 'int64'])
    z_scores = np.abs((numeric_cols - numeric_cols.mean()) / numeric_cols.std())

    return list(set(np.where(z_scores > threshold)[0]))

zscore_outliers = detect_outliers_zscore(df)
print("\nTotal rows with outliers (Z-score):", len(zscore_outliers))

# VISUAL OUTLIER CHECK (BOXPLOTS)

df.select_dtypes(include=['float64', 'int64']).plot(
    kind='box', subplots=True, figsize=(12, 8),
    layout=(3, 4), sharex=False, sharey=False, title="Boxplots for Outliers"
)
plt.tight_layout()
plt.show()

# INCONSISTENCY CHECKS

print("\nINCONSISTENCIES : ")
print("Duplicate Rows")
print("Number of duplicates:", df.duplicated().sum())
print("Data Types")
print(df.dtypes)

# Negative values check (for features where negatives are not valid)
print("\nNegative Values (Possible Errors) : ")
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    if (df[col] < 0).any():
        print(f"Column {col}: Contains negative values")
    else:
        print(f"Column {col}: OK")


# Out-of-range values (example: glass dataset ranges)
print("\nOut-of-Range Checks (Example for Glass Dataset) : ")
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    print(f"{col}: min={df[col].min()} | max={df[col].max()}")
    
#  SUMMARY

print("\nSUMMARY:")
print(f"Missing value columns: {(df.isnull().sum() > 0).sum()}")
print(f"Outliers (IQR): {len(iqr_outliers)} rows")
print(f"Outliers (Z-score): {len(zscore_outliers)} rows")
print(f"Duplicates: {df.duplicated().sum()} rows")
print("Negative value check completed.")
```

```python
# 3. Data Preprocessing
# 1. Check for missing values in the dataset and decide on a strategy for handling them.Implement the chosen strategy (e.g., imputation or removal) and explain your reasoning.

# CHECK MISSING VALUES

print("MISSING VALUES CHECK")
missing_count = df.isnull().sum()
missing_percent = (df.isnull().mean() * 100).round(2)

print("Missing Values Per Column:\n", missing_count)
print("\nMissing Percentage Per Column:\n", missing_percent)

# Summary
print("\nTotal columns with missing values:", (missing_count > 0).sum())

# DECIDE STRATEGY BASED ON MISSINGNESS

# Example strategy:
# - If missing < 5% â†’ Impute with mean/median (safe)
# - If missing 5â€“25% â†’ Advanced imputation (KNN, regression)
# - If missing > 25% â†’ Consider removing column or domain-specific decision

strategy = {}

for col in df.columns:
    pct = missing_percent[col]
    
    if pct == 0:
        strategy[col] = "No missing values"
    elif pct < 5:
        strategy[col] = "Impute with median (numeric) or mode (categorical)"
    elif pct < 25:
        strategy[col] = "Impute using advanced method (KNN) â€” optional"
    else:
        strategy[col] = "Consider dropping the column or careful review"

print("\nSELECTED STRATEGY PER COLUMN")
for col, decision in strategy.items():
    print(f"{col}: {decision}")

# IMPLEMENTATION OF CHOSEN STRATEGY
#    (Simple & commonly used approach)

df_clean = df.copy()

for col in df.columns:
    if missing_count[col] > 0:
        if df[col].dtype in ['float64', 'int64']:    # numeric column
            df_clean[col] = df[col].fillna(df[col].median())
        else:                                        # categorical column
            df_clean[col] = df[col].fillna(df[col].mode()[0])

print("\nMISSING VALUES AFTER CLEANING : ")
print(df_clean.isnull().sum())

# Save cleaned dataset
df_clean.to_excel("glass_cleaned.xlsx", index=False)
print("\nCleaned file saved as 'glass_cleaned.xlsx'")
```

### EXPLANATION OF THE STRATEGY

#### 1. Why Median Imputation for Numeric Features?

##### Median is robust to outliers, unlike mean

##### Suitable for skewed distributions (common in real data)


#### 2. Why Mode Imputation for Categorical Features?

##### Mode represents the most common category

##### Preserves the structure of the data


#### 3. When Should We Remove Columns?

##### Remove a column only if:

##### Missing values are more than 25â€“30%

##### The feature is not important

##### There is no reliable way to impute it


##### This is rare for the Glass dataset.

#### 4. Why Not Use KNN or Regression Imputation Here?

##### They are powerful but:

##### Require tuning

##### Can introduce bias if dataset is small

##### More complex for introductory-level analysis


##### Median/mode imputation is simpler and effective for most cases.

```python
# 3. Apply feature scaling techniques such as standardization or normalization to ensure that all features are on a similar scale. Handling the imbalance data.
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from imblearn.over_sampling import SMOTE

# SEPARATE FEATURES AND TARGET
#    (Assume 'Type' is the target column â€” change if needed)
target_col = "Type"   # Modify if your target column has a different name

X = df.drop(columns=[target_col])
y = df[target_col]

print("Features and target separated.\n")

# IDENTIFY NUMERICAL COLUMNS

numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
print("Numeric Columns:\n", numeric_cols)

# FEATURE SCALING
# ---------- Standardization (Z-score) ----------
scaler_standard = StandardScaler()
X_standard = X.copy()
X_standard[numeric_cols] = scaler_standard.fit_transform(X[numeric_cols])

print("\nStandardization Applied (mean=0, std=1).\n")

# ---------- Normalization (Min-Max) ----------
scaler_minmax = MinMaxScaler()
X_normalized = X.copy()
X_normalized[numeric_cols] = scaler_minmax.fit_transform(X[numeric_cols])

print("Normalization Applied (scaled to [0,1]).\n")

# HANDLE IMBALANCED DATA USING SMOTE

smote = SMOTE(random_state=42)

# Apply SMOTE on standardized data
X_std_balanced, y_std_balanced = smote.fit_resample(X_standard, y)

# Apply SMOTE on normalized data (optional)
X_norm_balanced, y_norm_balanced = smote.fit_resample(X_normalized, y)

print("SMOTE applied successfully â€” imbalance handled.\n")

# SAVE CLEANED OUTPUT FILES
# Standardized + SMOTE
std_df = pd.concat([pd.DataFrame(X_std_balanced, columns=X_standard.columns),
                    pd.DataFrame(y_std_balanced, columns=[target_col])],
                   axis=1)
std_df.to_excel("glass_standardized_balanced.xlsx", index=False)

# Normalized + SMOTE
norm_df = pd.concat([pd.DataFrame(X_norm_balanced, columns=X_normalized.columns),
                     pd.DataFrame(y_norm_balanced, columns=[target_col])],
                    axis=1)
norm_df.to_excel("glass_normalized_balanced.xlsx", index=False)

print("Files saved:\n- glass_standardized_balanced.xlsx\n- glass_normalized_balanced.xlsx")
```

```python
# 4. Random Forest Model Implementation
# 1. Divide the data into train and test split.
from sklearn.model_selection import train_test_split

# Divide data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2,      # 20% test data
    random_state=42,    # for reproducibility
    stratify=y          # keeps the target class distribution same in train & test
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

```python
# 2. Implement a Random Forest classifier using Python and a machine learning library like scikit-learn.
# RANDOM FOREST CLASSIFIER

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Create the Random Forest model
rf = RandomForestClassifier(
    n_estimators=200,        # number of trees
    max_depth=None,         # trees grow fully
    random_state=42,        # reproducibility
    class_weight='balanced' # handles imbalance
)

# Train the model
rf.fit(X_train, y_train)

# Predict on test data
y_pred = rf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# 3. Train the model on the train dataset. Evaluate the performance on test data using metrics like accuracy, precision, recall, and F1-score.
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Train the Random Forest Model

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'   # handles imbalanced data
)

rf_model.fit(X_train, y_train)

# Make Predictions

y_pred = rf_model.predict(X_test)

# Evaluate the Model

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')   # works for multiclass
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision (Weighted):", precision)
print("Recall (Weighted):", recall)
print("F1 Score (Weighted):", f1)

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# 5. Bagging and Boosting Methods

# Apply the Bagging and Boosting methods and compare the results.

from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Function to evaluate models

def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return accuracy, precision, recall, f1

# BAGGING CLASSIFIER

bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    bootstrap=True,
    random_state=42
)

# ADA BOOST (Boosting)

ada = AdaBoostClassifier(
    n_estimators=100,
    learning_rate=0.8,
    random_state=42
)

# GRADIENT BOOSTING

gb = GradientBoostingClassifier(
    n_estimators=150,
    learning_rate=0.05,
    random_state=42
)

# Optional XGBoost (only if installed)
try:
    from xgboost import XGBClassifier

    xgb = XGBClassifier(
        n_estimators=200,
        learning_rate=0.1,
        max_depth=4,
        random_state=42
    )

    use_xgb = True
except:
    use_xgb = False

# Evaluate All Models

models = {
    "Bagging": bagging,
    "AdaBoost": ada,
    "Gradient Boosting": gb
}

if use_xgb:
    models["XGBoost"] = xgb

results = {}

for name, model in models.items():
    print(f"\n{name} : ")
    accuracy, precision, recall, f1 = evaluate_model(model, X_train, y_train, X_test, y_test)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)

    results[name] = [accuracy, precision, recall, f1]

# Compare Results

import pandas as pd

results_df = pd.DataFrame(results, index=["Accuracy", "Precision", "Recall", "F1 Score"])
print("\nMODEL PERFORMANCE COMPARISIONS : ")
print(results_df)
```

## 1. Explain Bagging and Boosting methods. How is it different from each other.
* Bagging (Bootstrap Aggregating) and Boosting are both ensemble learning techniques used to improve the performance of machine learning models.
* Bagging works by training multiple models (usually of same type) independently using different subsets of the training data obtain through bootstrapping and then the final prediction is made by aggregating the predictions of all models (e.g., majority voting for classification or averaging for regression).
* Boosting, on the other hand, trains models sequentially, with each new model focusing on the errors made by the previous ones. This helps to reduce bias and improve overall accuracy.
* In summary, Bagging is parallel and reduces variance, while Boosting is sequential and reduces bias.

## 2. Explain how to handle imbalance in the data.
Handling imbalanced data is crucial for building effective machine learning models. Here are some common techniques:
1. Resampling Techniques:
    - Oversampling: Increase the number of instances in the minority class by duplicating examples or generating synthetic examples (e.g., SMOTE).
    - Undersampling: Reduce the number of instances in the majority class by randomly removing examples.
2. Use Different Evaluation Metrics:
    -  Instead of accuracy, use metrics like precision, recall, F1-score, and ROC-AUC that provide better insight into model performance on imbalanced data.
3. Algorithmic Approaches:
    - Use algorithms that are robust to class imbalance (e.g., tree-based methods).
    - Modify existing algorithms to incorporate class weights, giving more importance to the minority class.
4. Ensemble Methods:
    - Use ensemble techniques like Bagging and Boosting that can help improve performance on imbalanced datasets.
