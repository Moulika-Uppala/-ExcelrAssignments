#### 1.Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Initial list
nums = [10, 20, 30, 40, 50]
print("Original list:", nums)

# append
nums.append(60)
print("After append:", nums)

# extend
nums.extend([70, 80])
print("After extend:", nums)

# insert
nums.insert(2, 25)
print("After insert at index 2:", nums)

# remove
nums.remove(40)
print("After removing 40:", nums)

# pop
nums.pop()  # removes last element
print("After pop:", nums)

# index
print("Index of 30:", nums.index(30))

# count
print("Count of 20:", nums.count(20))

# sort
nums.sort()
print("After sorting:", nums)

# reverse
nums.reverse()
print("After reversing:", nums)

# clear
nums.clear()
print("After clear:", nums)
```

#### 7.Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative. 

```python
num = int(input("Enter a number: "))

if num > 0:
    if num % 2 == 0:
        print("The number is positive and even.")
    else:
        print("The number is positive but odd.")
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

### Section C: Exploring Pandas

#### 11.Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
import pandas as pd
import numpy as np


# 1. Create a DataFrame manually


data = {
    "Name": ["Amit", "Riya", "John", "Amit", "Sara", "Tom", "Neha", "Riya", "David", "John"],
    "Age": [25, 30, 22, 25, 28, 35, 27, 30, 40, 22],
    "City": ["Delhi", "Mumbai", "Kolkata", "Delhi", "Chennai", "Delhi", "Pune", "Mumbai", "Kolkata", "Kolkata"],
    "Salary": [50000, 60000, np.nan, 50000, 55000, 65000, 48000, 60000, np.nan, 70000]
}

df = pd.DataFrame(data)

print("\n Original DataFrame ")
print(df)


# 2. Use info() and describe()

print("\n DataFrame Info ")
print(df.info())

print("\n Describe ")
print(df.describe())


# 3. Select Name and City columns


print("\n Name & City Columns ")
print(df[["Name", "City"]])


# 4. Drop the City column

df = df.drop("City", axis=1)
print("\n After Dropping City Column ")
print(df)


# 5. Fill missing Salary with mean


mean_salary = df["Salary"].mean()
df["Salary"].fillna(mean_salary, inplace=True)

print("\n After Filling Missing Salary with Mean ")
print(df)


# 6. Remove duplicate rows


df = df.drop_duplicates()

print("\n Final DataFrame After Removing Duplicates ")
print(df)
```
