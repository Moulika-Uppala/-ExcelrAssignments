```python
### Task 2: Dimensionality Reduction with PCA:
# 1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load dataset

df=pd.read_csv("C:\\Users\\moulika\\Downloads\\wine.csv")

# Separate features (X) and target label (y)
X = df.drop("Type", axis=1)   # independent variables
y = df["Type"]                # target variable

# 2. Standardize the data

# StandardScaler transforms each feature:
#   - mean = 0
#   - standard deviation = 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Standardized Data (first 5 rows):\n")
print(pd.DataFrame(X_scaled, columns=X.columns).head())

# 3. Apply PCA

# n_components=None → PCA keeps all components but will tell how many matter.
pca = PCA(n_components=None)
X_pca = pca.fit_transform(X_scaled)

# 4. Explained variance

print("\nExplained Variance Ratio (per component):\n")
print(pca.explained_variance_ratio_)

print("\nCumulative Explained Variance:\n")
print(pca.explained_variance_ratio_.cumsum())

# 5. Convert PCA results into a DataFrame

pca_df = pd.DataFrame(
    X_pca,
    columns=[f"PC{i+1}" for i in range(X_pca.shape[1])]
)

print("\nPCA-Transformed Data (first 5 rows):\n")
print(pca_df.head())
```

```python
# 2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load and standardize the dataset
df=pd.read_csv("C:\\Users\\moulika\\Downloads\\wine.csv")

X = df.drop("Type", axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. PCA (keep all components)

pca = PCA()
pca.fit(X_scaled)

# 3. Scree Plot (Eigenvalues = Explained Variance)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_) + 1),
         pca.explained_variance_,
         marker='o')

plt.title("Scree Plot (Eigenvalues)")
plt.xlabel("Principal Component")
plt.ylabel("Eigenvalue (Explained Variance)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# 4. Cumulative Explained Variance Plot

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         pca.explained_variance_ratio_.cumsum(),
         marker='o')

plt.title("Cumulative Explained Variance Plot")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance Ratio")
plt.grid(True, linestyle="--", alpha=0.5)
plt.axhline(y=0.90, color='red', linestyle='--')  # mark 90% threshold
plt.text(1, 0.91, "90% Threshold", color='red')
plt.show()

# 5. Print exact explained variance values

print("Explained Variance Ratio per Component:\n")
print(pca.explained_variance_ratio_)

print("\nCumulative Explained Variance:\n")
print(pca.explained_variance_ratio_.cumsum())
```

```python
# 2.	Discuss any similarities or differences observed in the clustering results.

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA

# 1. Load dataset

df=pd.read_csv("C:\\Users\\moulika\\Downloads\\wine.csv")

X = df.drop("Type", axis=1)

# 2. Standardize

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. K-Means on ORIGINAL data

kmeans_original = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans_original.fit_predict(X_scaled)

sil_original = silhouette_score(X_scaled, clusters_original)
db_original = davies_bouldin_score(X_scaled, clusters_original)

# 4. PCA transformation

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 5. K-Means on PCA data

kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)

sil_pca = silhouette_score(X_pca, clusters_pca)
db_pca = davies_bouldin_score(X_pca, clusters_pca)

# 6. PRINT SIMILARITIES & DIFFERENCES

print("\n CLUSTERING COMPARISON: SUMMARY \n")

print(f"Silhouette Score (Original Data): {sil_original:.4f}")
print(f"Silhouette Score (PCA Data):      {sil_pca:.4f}\n")

print(f"Davies–Bouldin Index (Original): {db_original:.4f}")
print(f"Davies–Bouldin Index (PCA):      {db_pca:.4f}\n")


# INTERPRETATION SECTION
print("INTERPRETATION \n")

# Silhouette comparison
if sil_pca > sil_original:
    print("PCA clustering has a higher Silhouette Score, indicating better cluster separation.")
elif sil_pca < sil_original:
    print("Original dataset clustering has a higher Silhouette Score, meaning better-defined clusters.")
else:
    print("Both methods produce equal Silhouette Scores.")

# Davies–Bouldin comparison
if db_pca < db_original:
    print("PCA clustering has a lower Davies–Bouldin Index → tighter and more separated clusters.")
elif db_pca > db_original:
    print("Original data has a lower Davies–Bouldin Index → better clustering compactness.")
else:
    print("Davies–Bouldin Index is equal for both methods.")

# General observation
print("\nOverall Observation:")
print("PCA reduces dimensionality and may improve clustering if noise or redundant features exist.")
print("If PCA performs slightly worse, it may indicate that the original high-dimensional structure")
print("contained important information that PCA compressed away.\n")

print("This summary helps describe the similarities and differences in assignment.")
```

```python
# 3.	Reflect on the impact of dimensionality reduction on clustering performance.

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score

# 1. Load dataset

df=pd.read_csv("C:\\Users\\moulika\\Downloads\\wine.csv")
X = df.drop("Type", axis=1)

# 2. Standardize data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. KMeans on original data

kmeans_orig = KMeans(n_clusters=3, random_state=42)
clusters_orig = kmeans_orig.fit_predict(X_scaled)

sil_orig = silhouette_score(X_scaled, clusters_orig)
db_orig = davies_bouldin_score(X_scaled, clusters_orig)

# 4. PCA Transformation

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 5. KMeans on PCA-reduced data

kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)

sil_pca = silhouette_score(X_pca, clusters_pca)
db_pca = davies_bouldin_score(X_pca, clusters_pca)

# 6. Automatic Reflection / Interpretation

print("\nIMPACT OF PCA ON CLUSTERING PERFORMANCE\n")

print(f"Silhouette Score (Original Data): {sil_orig:.4f}")
print(f"Silhouette Score (PCA Data):      {sil_pca:.4f}\n")

print(f"Davies-Bouldin Index (Original): {db_orig:.4f}")
print(f"Davies-Bouldin Index (PCA):      {db_pca:.4f}\n")

print("INTERPRETATION \n")

# Silhouette interpretation
if sil_pca > sil_orig:
    print("PCA improved the Silhouette Score, indicating that dimensionality reduction helped create clearer and more well-separated clusters.")
else:
    print("The Silhouette Score decreased after applying PCA, suggesting that dimensionality reduction may have removed useful variance needed for clustering.")

# Davies–Bouldin interpretation
if db_pca < db_orig:
    print("The lower Davies–Bouldin Index after PCA indicates tighter and more compact clusters in lower dimensions.")
else:
    print("The Davies–Bouldin Index increased after PCA, meaning clusters became less compact and less separated.")

# Overall reflection
print("\nOverall Reflection:")
print("Dimensionality reduction with PCA can improve clustering when high-dimensional data contains noise or redundant features.")
print("However, if PCA reduces important variance, clustering may perform worse. The impact depends on how much discriminative information PCA preserves in fewer components.\n")
```

```python
# 4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score
import time

# 1. Load dataset

df=pd.read_csv("C:\\Users\\moulika\\Downloads\\wine.csv")
X = df.drop("Type", axis=1)

# 2. Standardize original data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. K-Means on ORIGINAL data

start_orig = time.time()
kmeans_orig = KMeans(n_clusters=3, random_state=42)
clusters_orig = kmeans_orig.fit_predict(X_scaled)
end_orig = time.time()

sil_orig = silhouette_score(X_scaled, clusters_orig)
db_orig = davies_bouldin_score(X_scaled, clusters_orig)
time_orig = end_orig - start_orig

# 4. PCA Transformation (retain 95% variance)

pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)
variance_retained = sum(pca.explained_variance_ratio_)

# 5. K-Means on PCA data

start_pca = time.time()
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)
end_pca = time.time()

sil_pca = silhouette_score(X_pca, clusters_pca)
db_pca = davies_bouldin_score(X_pca, clusters_pca)
time_pca = end_pca - start_pca

# 6. Print side-by-side comparison

print("\n----------------- PERFORMANCE COMPARISON -----------------\n")

print(f"Original Data Shape:        {X_scaled.shape}")
print(f"PCA Reduced Shape:          {X_pca.shape}")
print(f"Variance Retained by PCA:   {variance_retained:.4f}\n")

print("Silhouette Score (Higher = Better):")
print(f" - Original Data:           {sil_orig:.4f}")
print(f" - PCA Data:                {sil_pca:.4f}\n")

print("Davies–Bouldin Index (Lower = Better):")
print(f" - Original Data:           {db_orig:.4f}")
print(f" - PCA Data:                {db_pca:.4f}\n")

print("Clustering Runtime (seconds):")
print(f" - Original Data:           {time_orig:.6f}")
print(f" - PCA Data:                {time_pca:.6f}\n")

# 7. Automatic trade-off analysis

print("-------------------- TRADE-OFF ANALYSIS --------------------\n")

# Dimensionality impact
print(f"PCA reduced dimensionality from {X_scaled.shape[1]} → {X_pca.shape[1]} features.")
print("This reduces computational cost and may remove noise.\n")

# Performance trade-offs
if sil_pca > sil_orig:
    print("PCA improved the Silhouette Score → better separation of clusters.")
else:
    print("PCA reduced the Silhouette Score → clustering became less distinct.")

if db_pca < db_orig:
    print("PCA lowered the Davies–Bouldin Index → clusters became tighter and more compact.")
else:
    print("PCA increased the Davies–Bouldin Index → clusters became less compact.")

# Speed trade-off
if time_pca < time_orig:
    print("\n PCA significantly reduced computation time for K-Means.")
else:
    print("\n PCA did not reduce computation time, possibly due to overhead.")

print("\n------------------ OVERALL INTERPRETATION ------------------\n")
print("Using PCA simplifies the dataset by removing redundancy and noise, which can improve clustering")
print("performance *if* the removed dimensions were not useful. However, if PCA eliminates important")
print("variance, clustering on PCA data may perform worse. The choice depends on whether simplicity and")
print("speed outweigh the loss of information from dimensionality reduction.\n")
```
